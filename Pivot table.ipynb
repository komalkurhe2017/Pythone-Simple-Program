```python
import pandas as pd
import numpy as np
```

```python
mydis={'Date':['5/1/2017','5/2/2017','5/3/2017','5/1/2017','5/2/2017','5/3/2017','5/1/2017','5/2/2017'],
       'city':['Shrirampur','Shrirampur','Belapur','Shrirampur','Belapur','Belapur','Nasik','Nasik'],
       'temp':[65,66,68,75,78,82,80,77],
       
       'humidity':[56,58,60,80,83,85,26,30]
      }
df=pd.DataFrame(mydis)
df
```

```python
df.info()
```

```python
df.isnull().any()
```

```python
df.pivot_table(index='Date')
```

```python
df.pivot_table(index='city')
```

```python
df
```

```python
df.pivot_table(index=['Date','city'])
```

```python
df.pivot_table(index='city',columns='Date')
```

```python
df.pivot_table(index='Date',columns='city')
```

```python
df.pivot_table(index='city',aggfunc=['sum','mean'])
```

```python
df.pivot_table(index='Date',columns='city',fill_value=22)
```

```python
t1=int(input("Enter the fill value : " ))
Data=df.pivot_table(index='Date',columns='city',fill_value=t1)
Data
```
